*–ú–∞–ª–∞—è –¥–æ–ª—è –ø—Ä–æ–ø—É—Å–∫–æ–≤ (<2%)*. –î–ª—è —Å—Ç–æ–ª–±—Ü–æ–≤ `name`, `year_of_release`, `genre` —Ä–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç—Å—è —É–¥–∞–ª–∏—Ç—å —Å—Ç—Ä–æ–∫–∏ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏, —Ç–∞–∫ –∫–∞–∫ –∏—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–µ–∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ. –≠—Ç–æ –ø—Ä–µ–¥–æ—Ç–≤—Ä–∞—Ç–∏—Ç –≤–ª–∏—è–Ω–∏–µ –Ω–µ–∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω—ã—Ö –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –∏—Ç–æ–≥–æ–≤—ã–π –∞–Ω–∞–ª–∏–∑.

*–°—Ä–µ–¥–Ω–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –ø—Ä–æ–¥–∞–∂ `(eu_sales, jp_sales)`*.–≠—Ç–∏ —Å—Ç–æ–ª–±—Ü—ã –∏–º–µ—é—Ç –º–∞–ª–æ–µ —á–∏—Å–ª–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ `(~–º–µ–Ω–µ–µ 2%)` —Ä–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç—Å—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –º–µ–¥–∏–∞–Ω—É –¥–ª—è –∑–∞–ø–æ–ª–Ω–µ–Ω–∏—è –ø—Ä–æ–ø—É—Å–∫–æ–≤, –ø–æ—Å–∫–æ–ª—å–∫—É, –æ–Ω–∞ —É—Å—Ç–æ–π—á–∏–≤–∞ –∫ –≤—ã–±—Ä–æ—Å–∞–º, –ª—É—á—à–µ –æ—Ç—Ä–∞–∂–∞–µ—Ç —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω—É—é —Ç–µ–Ω–¥–µ–Ω—Ü–∏—é —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.

*–î–ª—è –¥—Ä—É–≥–∏—Ö —Å—Ç–æ–ª–±—Ü–æ–≤* —Å –±–æ–ª—å—à–∏–º —á–∏—Å–ª–æ–º –ø—Ä–æ–ø—É—Å–∫–æ–≤ —Å–ª–µ–¥—É–µ—Ç –ø—Ä–æ–≤–µ—Å—Ç–∏ –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ–µ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ –ø–µ—Ä–µ–¥ –ø—Ä–∏–Ω—è—Ç–∏–µ–º —Ä–µ—à–µ–Ω–∏—è –æ –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–∏ –∏–ª–∏ —É–¥–∞–ª–µ–Ω–∏–∏ –ø—Ä–æ–ø—É—Å–∫–æ–≤, —Ç–∞–∫ –∫–∞–∫ –∏–≥–Ω–æ—Ä–∏—Ä–æ–≤–∞–Ω–∏–µ –∏–ª–∏ —É–¥–∞–ª–µ–Ω–∏–µ –±–æ–ª—å—à–∏—Ö –æ–±—ä–µ–º–æ–≤ –¥–∞–Ω–Ω—ã—Ö –º–æ–∂–µ—Ç —Å—É—â–µ—Å—Ç–≤–µ–Ω–Ω–æ –ø–æ–≤–ª–∏—è—Ç—å –Ω–∞ –∫–∞—á–µ—Å—Ç–≤–æ –∞–Ω–∞–ª–∏–∑–∞.


–ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –∫–æ—Ç–æ—Ä–∞ –Ω–µ–ø–æ—Å—Ä–¥–µ—Å–≤—Ç–µ–Ω–Ω–æ –ø–æ–º–æ–≥–∞–µ—Ç –æ–±–Ω–∞—Ä—É–∂–∏–≤–∞—Ç—å –∏  –∑–∞–ø–æ–ª–Ω—è—Ç—å –ø—Ä–æ–ø—É—Å–∫–∏ –∏—Å—Ö–æ–¥—è –∏–∑ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤.


```python
##–ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –∫–æ—Ç–æ—Ä–∞—è –Ω–µ–ø–æ—Å—Ä–µ–¥—Å—Ç–≤–µ–Ω–Ω–æ –ø–æ–º–æ–≥–∞–µ—Ç –æ–±–Ω–∞—Ä—É–∂–∏–≤–∞—Ç—å –∏  –∑–∞–ø–æ–ª–Ω—è—Ç—å –ø—Ä–æ–ø—É—Å–∫–∏ –∏—Å—Ö–æ–¥—è –∏–∑ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤.
def fill_missing(row, col_name):
    if pd.isna(row[col_name]):
        group= (games['platform'] == row['platform']) & (games['year_of_release'] == row['year_of_release'])
        mean_val = games[group][col_name].median()
        return mean_val
    else:
        return row[col_name]

# –ü—Ä–∏–º–µ–Ω–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –∑–∞–ø–æ–ª–Ω–µ–Ω–∏—è –ø—Ä–æ–ø—É—Å–∫–æ–≤
for col in ['eu_sales', 'jp_sales']:
    games[col] = games.apply(lambda row: fill_missing(row, col), axis=1)
```
